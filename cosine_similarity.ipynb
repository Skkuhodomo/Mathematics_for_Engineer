# **코사인 유사도(cosine similarity)**

lagrida_latex_editor-2.png



수식은 이름 그대로, 코사인 값이다. 내적의 정의를 모른다면, 고등교육과정 '기하' 또는 대학 미분적분학을 학습하는 걸 추천한다.

상관 계수 측정과 달리, 불연속, 비연관 데이터의 유사도 측정에 유리하다.


예시) matpoltlib을 이용해 그린 그래프를 통해여 직관적으로 이해할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 두 벡터 생성
vector_a = np.array([1, 2])  # [x1, y1]
vector_b = np.array([3, 4])  # [x2, y2]

# 벡터를 그래프 상에 시각화
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
plt.quiver(0, 0, vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')

# 벡터 간의 각도를 계산
cosine_similarity = np.dot(vector_a, vector_b) / (np.linalg.norm(vector_a) * np.linalg.norm(vector_b))
theta = np.arccos(cosine_similarity) * (180 / np.pi)  # 라디안을 도로 변환

# 각도 정보를 표시
mid_point = (vector_a + vector_b) / 2
plt.text(mid_point[0], mid_point[1], f'Theta = {theta:.2f} degrees\nCosine Similarity = {cosine_similarity:.2f}', fontsize=12)

# 그래프 설정
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.title('Cosine Similarity Visualization')
plt.legend()
plt.show()
```

# **파이썬에서 코사인 유사도**

복잡한 수식을 매번 증명할 필요도 없을 뿐더러, 수식을 작성할 필요 역시 없다.

다음과 같은 방법으로 사용하면 된다.



```python
from sklearn.metrics.pairwise import cosine_similarity
vector_a = vector_a.reshape(1, -1)
vector_b = vector_b.reshape(1, -1)
result = cosine_similarity(vector_a, vector_b)
result
```
